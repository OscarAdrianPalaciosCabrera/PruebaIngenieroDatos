```python
%sql
--A continuacion vamos a optimizar el las tipicas consultas donde se filtra por cierta fecha o cierto rango de fechas y con SQL tradicional se leen todos los archivos, icluso los que no aplican a el rango de fechas que estamos buscando

--Vamos a reorganizar los datos por trip_date
OPTIMIZE workspace.gold.hvfhs_kpis_diarios
ZORDER BY (trip_date);

```

```python
%sql
-- Ahora ya podemos realizar una consulta donde se reduce significativamente la cantidad de archivos leidos y la latencia
--debido a que ya no se gastara memoria leyendo archivos que no contienen los datos que estamos buscando 
EXPLAIN
SELECT *
FROM workspace.gold.hvfhs_kpis_diarios
WHERE trip_date = '2025-01-10';

--en el explain se puede ver Datafilters: en donde el filtro WHERE 2025-01-10 se hace antes de leer, no despues de leer, es decir que se realizo dataskipping 
```
